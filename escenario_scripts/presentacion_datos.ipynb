Exploracion de los datos obtenidos de la primeras simulaciones

```python
#importamos las librerias y recursos
import pandas as pd 
import seaborn as sns 

#Leemos el csv
df = pd.read_csv("../datos_escenario/15_mincap.csv",index_col="No.")

#Mostramos los primeros 30 paquetes de la captura
display(df.head(30))
```

```python
group_df = df.groupby("Protocol").size()

ax = sns.barplot(x=group_df.index, y=group_df)
```

```python
#Obtenemos las features consideradas y las dividimos según los protocolos

features = df.columns

general1_feat = features[:6].tolist()

general2_feat = features[21:26].tolist()

opc_feat = features[6:10].tolist()

s7_feat = features[10:15].tolist()

mod_feat = features[15:21].tolist()

print("Las caracteristicas generales son: " + str(general1_feat + general2_feat ))

print("Las caracteristicas del protocolo OPC UA son: " + str(opc_feat))

print("Las caracteristicas del protocolo S7 son: " + str(s7_feat))

print("Las caracteristicas del protocolo Modbus/TCP son: " + str(mod_feat))
```

```python
#Podemos obtener un dataFrame(tabla) por protocolo

opc_columns = general1_feat + general2_feat + opc_feat

opc_df = df[opc_columns][(df.Protocol=="OpcUa") | (df.Protocol=="TCP")]

opc_df
```

```python
#Podemos obtener un dataFrame(tabla) por protocolo


s7_columns = general1_feat + general2_feat + s7_feat

s7_df = df[s7_columns][df.Protocol=="S7COMM"]

s7_df
```

```python
#Podemos obtener un dataFrame(tabla) por protocolo


mod_columns = general1_feat + general2_feat + mod_feat

mod_df = df[mod_columns][(df.Protocol=="Modbus/TCP")]

mod_df
```
